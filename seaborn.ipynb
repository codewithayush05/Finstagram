```python
"""Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. They help identify patterns, correlations, and outliers in data.

Bar Plot: Bar plots are used to display and compare categorical data. They are suitable for showing the distribution of a single categorical variable or comparing multiple categories.

Box Plot: Box plots are useful for visualizing the distribution and spread of numerical data. They show the median, quartiles, and potential outliers in the data.

Heatmap: Heatmaps are used to visualize the relationships between two categorical variables. They are particularly helpful for displaying correlation matrices and complex datasets.

Line Plot: Line plots are used to represent data points connected by lines. They are suitable for visualizing trends and changes in data over a continuous variable, often used for time series data."""
```

```python
flights_data = sns.load_dataset("flights")
```

```python
flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")


plt.figure(figsize=(10, 6))
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)
```
