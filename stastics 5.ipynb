```python
import scipy.stats as stats

mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 100 

margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (std_dev / (sample_size ** 0.5))
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

```python
import scipy.stats as stats

observed_counts = [25, 18, 22, 12, 11, 12]

expected_counts = [
    0.20 * sum(observed_counts),  
    0.20 * sum(observed_counts),  
    0.20 * sum(observed_counts),  
    0.10 * sum(observed_counts),  
    0.10 * sum(observed_counts),  
    0.20 * sum(observed_counts)   
]

chi_square, p_value = stats.chisquare(observed_counts, f_exp=expected_counts)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The distribution of M&M colors is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution of M&M colors is not significantly different from the expected distribution.")
```

```python
import scipy.stats as stats

sample_size = 500
smokers = 60

confidence_level = 0.95
standard_error = (smokers / sample_size) ** 0.5

z_score = stats.norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_score * standard_error

lower_bound = (smokers / sample_size) - margin_of_error
upper_bound = (smokers / sample_size) + margin_of_error

print("95% Confidence Interval for Proportion of Smokers: ({:.4f}, {:.4f})".format(lower_bound, upper_bound))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Define the range of chi-square values
x = np.linspace(0, 30, 1000)

# Calculate the chi-square probability density function (PDF)
pdf = stats.chi2.pdf(x, df)

# Create the chi-square distribution plot
plt.plot(x, pdf, label=f'Chi-Square (df={df})', color='b')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 100)
y_fill = stats.chi2.pdf(x_fill, df)
plt.fill_between(x_fill, 0, y_fill, alpha=0.2, color='r', label='Chi-Square < 15')

# Label axes and add legend
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Chi-Square Distribution with Shaded Area (df=10)')
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

sample_size = 1000
coke_preference = 520

confidence_level = 0.99
standard_error = (coke_preference / sample_size) ** 0.5

z_score = stats.norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_score * standard_error

lower_bound = (coke_preference / sample_size) - margin_of_error
upper_bound = (coke_preference / sample_size) + margin_of_error

print("99% Confidence Interval for Proportion Preferring Coke: ({:.4f}, {:.4f})".format(lower_bound, upper_bound))
```

```python
import scipy.stats as stats

observed_tails = 45
observed_heads = 100 - observed_tails
expected_tails = 50  # Expected number of tails for a fair coin
expected_heads = 50  # Expected number of heads for a fair coin

observed_counts = [observed_tails, observed_heads]
expected_counts = [expected_tails, expected_heads]

chi_square, p_value = stats.chisquare(observed_counts, f_exp=expected_counts)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The coin may be biased towards tails.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence of bias towards tails.")
```

```python
import scipy.stats as stats

# Create a contingency table
observed = [[20, 15], [10, 25], [15, 20]]

chi_square, p_value, _, _ = stats.chi2_contingency(observed)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between smoking status and lung cancer diagnosis.")
```

```python
import scipy.stats as stats

sample_size = 30
sample_mean = 72
sample_std_dev = 10
population_mean = 70

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

alpha = 0.05
degrees_of_freedom = sample_size - 1
critical_value = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)

if abs(t_statistic) > critical_value:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the population mean and 70.")
```
