```python
from bokeh.plotting import figure, show


p = figure(title="Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis")

p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color="blue", legend_label="Circle Glyph")
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2, color="red", legend_label="Line Glyph")

p.legend.title = "Glyph Types"

# Show the plot
show(p)
```
